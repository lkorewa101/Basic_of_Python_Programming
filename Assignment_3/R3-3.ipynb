컴퓨터소프트웨어학부 고기현

```python
strings = ['1', '2', '3', '4', '5', '6']
print('::'.join(strings))
```

```python
# 정수형 배열의 경우는 오류 발생
numbers = [1, 2, 3, 4, 5, 6]
print('::'.join(numbers))
```

```python
# numbers 배열의 각 요소를 str()함수를 사용해서 문자열로 변환이 필요

numbers = [1, 2, 3, 4, 5, 6]
print('::'.join(map(lambda x: str(x),numbers)))
```
