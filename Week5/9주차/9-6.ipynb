```python
#사전검사 while-반복문
total = 0
i = 0
while i < 5:
  num = float(input("숫자를 입력하시오: "))
  total = total + num
  i += 1
  avg = total / 5
print("총합: ", total, "평균: ", avg)
```

```python
#사후검사 while-반복문
total = 0
i = 0
while True:
  num = float(input("숫자를 입력하시오: "))
  total = total + num
  i += 1
  if i > 4: break
avg = total / 5
print("총합: ", total, "평균: ", avg)
```
