```python
def hi_korean(someone):
  print(f'안녕하세요, {someone}')

def hi_english(someone):
  print(f'Hello, {someone}')

def get_greeting(language):
  if language == 'K':
    return hi_korean
  else:
    return hi_english

hello = get_greeting('K')
hello('사과')

hello = get_greeting('E')
hello('Apple')
```
