컴퓨터소프트웨어학부 고기현

```python
month = int(input("월에 해당하는 정수를 입력하시오: "))
```

```python
if month == 12 or month == 1 or month == 2:
  print(month, "월은 겨울입니다.")
elif month == 3 or month == 4 or month == 5:
  print(month, "월은 봄입니다.")
elif month == 6 or month == 7 or month == 8:
  print(month, "월은 여름입니다.")
else:
  print(month, "월은 가을입니다.")
```
