```python
with open('calculator.py', 'w') as file:
  file.write(
"""
__version = "0.0.1"

def min(*numbers):
  minimum = numbers[0]
  for number in numbers[1:]:
    if number < minimum:
        minimum = number
  return minimum

def max(*numbers):
  maximum = numbers[0]
  for number in numbers[1:]:
    if number > maximum:
        maximum = number
  return maximum

""")
```
